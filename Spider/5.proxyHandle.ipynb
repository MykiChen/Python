```python
from urllib import request
from urllib import parse
```

```python
url = 'http://httpbin.org/ip'
resp = request.urlopen(url)
resp.read()
```

```python
url = 'http://httpbin.org/ip'
handler = request.ProxyHandler({'https':'117.57.91.2:39344'})
opener = request.build_opener(handler)
resp = opener.open(url)
resp.read()
```

```python
#get data from UTAS
url = 'https://www.utas.edu.au/postgraduate-study'

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.79 Safari/537.36',
}

req = request.Request(url, headers = headers)
resp = request.urlopen(req)
result = parse.urlparse(url)
print(result)
resp.read()
```
