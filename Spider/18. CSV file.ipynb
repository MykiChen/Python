```python
with open('abalone.csv', 'r') as fp:
    reader = csv.reader(fp)
    #next()
    next(reader)
    for x in reader:
        #print(x)
        sex = x[0]
        shell_weight = x[-2]
        print({'sex': sex, 'shell_weight': shell_weight})
        #print(shell_weight)
```

```python
with open('abalone.csv', 'r') as fp:
    #DictReader()
    reader = csv.DictReader(fp)
    for x in reader:
        #print(x['sex'])
        #print(x['length'])
        values = {'sex': x['sex'], 'length': x['length']}
        print(values)
```

```python
headers = [
    'sex', 'shell_weight', 'rings'
]
values = [
    ('F', 23, 4),
    ('M', 28, 3),
    ('F', 22, 6)
]

with open('abalone2.csv', 'w', encoding = 'utf-8', newline='') as fp:
    writer = csv.writer(fp)
    writer.writerow(headers)
    writer.writerows(values)
```

```python
def write_csv_demo2():
    headers = ['Name', 'age', 'height']
    values = [
        {"Name":'Chen', 'age': 23, 'height':124},
        {"Name":'Myki', 'age': 21, 'height':164},
        {"Name":'Xixi', 'age': 43, 'height':174}
    ]
    with open('class1.csv', 'w', encoding='utf-8', newline='') as fp:
        writer = csv.DictWriter(fp, headers)
        writer.writeheader()
        writer.writerows(values)
        
if __name__ == '__main__':
    write_csv_demo2()
```
