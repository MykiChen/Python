```python
import re
import requests
```

```python
def parse_page(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36'
    }
    response = requests.get(url, headers = headers)
    text = response.text
    titles = re.findall(r'<div\sclass="cont">.*?<b>(.*?)</b>', text, re.DOTALL)
    years = re.findall(r'<p\sclass="source"><a.*?>(.*?)</a>', text, re.DOTALL)
    authors = re.findall(r'<p\sclass="source"><a.*?>.*?</a>.*?<a.*?>(.*?)</a>', text, re.DOTALL)
    Contents = re.findall(r'<div\sclass="contson".*?>(.*?)</div>', text, re.DOTALL)
    
    poem_content = []
    for content in Contents:
        content = re.sub(r'<.*?>', "", content)
        #print(content.strip())
        poem_content.append(content.strip())
    
    poems = []
    for details in zip(titles, years, authors, poem_content):
        titles, years, authors, poem = details
        poem = {
            'Title': titles,
            'Dynasty': years,
            'Author': authors,
            'Content': poem
        }
        poems.append(poem)
        
    for poem in poems:
        print(poem)
        print('-'*30)
    
def main():
    base_url = "https://www.gushiwen.org/default_{}.aspx"
    for i in range(1,5):
        url = base_url.format(i)
        parse_page(url)
    
if __name__ == '__main__':
    main()
```
