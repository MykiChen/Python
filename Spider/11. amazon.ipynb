```python
import requests
from lxml import etree
```

```python
Headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.79 Safari/537.36',
    'Referer': 'https://www.amazon.cn/s?k=%E5%8D%95%E5%8F%8D&rh=n%3A755655051&__mk_zh_CN=%E4%BA%9A%E9%A9%AC%E9%80%8A%E7%BD%91%E7%AB%99&ref=nb_sb_noss'    
}

def get_details(url):
    response = requests.get(url, headers=Headers)
    text=response.content.decode('utf-8')
    html=etree.HTML(text)

    divs = html.xpath("//div[@class='a-section a-spacing-medium']")

    #cameras = []
    for div in divs:
        name = div.xpath(".//span[@class='a-size-medium a-color-base a-text-normal']//text()")[0]
        print('Name: ' + name)
        price = div.xpath(".//span[@class='a-offscreen']//text()")[0]
        print('Price: ' + price)
        href = div.xpath(".//div[@class='a-section a-spacing-none']//@href")[0]
        href = 'https://www.amazon.cn' + href
        print('Shopping Address: ' + href) 
        print("")

def get_more_cameras():
    base_url = 'https://www.amazon.cn/s?k=%E4%BD%B3%E8%83%BD&i=photo-video&rh=n%3A755655051&page={}&__mk_zh_CN=%E4%BA%9A%E9%A9%AC%E9%80%8A%E7%BD%91%E7%AB%99&qid=1576741208&ref=sr_pg_3'
    for i in range(2, 10):
        url = base_url.format(i)
        get_details(url)
    
if __name__ == '__main__':
    url = 'https://www.amazon.cn/s?k=%E4%BD%B3%E8%83%BD&rh=n%3A755655051&__mk_zh_CN=%E4%BA%9A%E9%A9%AC%E9%80%8A%E7%BD%91%E7%AB%99&ref=nb_sb_noss'
    get_details(url)
    get_more_cameras()
```
