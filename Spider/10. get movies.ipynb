```python
Headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.79 Safari/537.36',
    'Referer': 'https://www.dygod.net/html/gndy/index_2.html'
}

def get_details_urls(url):
    response = requests.get(url, headers=Headers)
    text = response.content.decode('gbk')
    html = etree.HTML(text)
    tables = html.xpath("//table[@class='tbspan']")

    for table in tables:
        #get details 
        type = table.xpath(".//a//text()")[0]
        title = table.xpath(".//a//text()")[1]
        href = table.xpath(".//a/@href")[1]
        href = 'https://www.dygod.net' + href
        date = table.xpath(".//font//text()")[0] 
        
        #get download link
        response_more = requests.get(href, headers=Headers)
        text_more = response_more.content.decode('gbk')
        html_more = etree.HTML(text_more)
        divs = html_more.xpath("//div[@class='co_content8']")
        download = divs[0].xpath(".//a[1]/@href")[0]
        downloadLink = 'Download Link: https://www.dygod.net/html/gndy/dyzz/' +  download + '\n'
    
        print(type + " " + title + " " + date)
        print(href)
        print(downloadLink)
        
def get_more_movies():
    base_url = 'https://www.dygod.net/html/gndy/china/index_{}.html'
    for i in range(2,8):
        url = base_url.format(i) 
        get_details_urls(url)
        
if __name__ == '__main__':
    url = 'https://www.dygod.net/html/gndy/china/index.html'
    get_details_urls(url)   
    get_more_movies()
    
```

```python
url = 'https://www.dygod.net/html/gndy/dyzz/20191204/112694.html'
response = requests.get(url, headers=Headers)

text = response.content.decode('gbk')
html = etree.HTML(text)
divs = html.xpath("//div[@id='Zoom']")[0]

infos = divs.xpath(".//text()")

for index,info in enumerate(infos):
    if info.startswith("◎译　　名"):
        info = info.replace("◎译　　名", "").strip()
        print('Name: ' + info)
    elif info.startswith("◎片　　名"):
        info = info.replace("◎片　　名", "").strip()
        print('English Name: ' + info)
    elif info.startswith("◎上映日期"):
        info = info.replace("◎上映日期", "").strip()
        print('Date: ' + info)
    elif info.startswith("◎豆瓣评分"):
        info = info.replace("◎豆瓣评分", "").strip()
        print('Score: ' + info)
    elif info.startswith("◎主　　演")   :
        info = info.replace("◎主　　演", " ").strip()
        print('Actors: ')
        print(info)
        for x in range(index+1, len(infos)):
            actor = infos[x].strip()
            if actor.startswith("◎简　　介"):
                break
            print(actor)
        print("")
    elif info.startswith("◎简　　介"):
        info = info.replace("◎简　　介", "").strip()
        print('Introduction: ')
        for x in range(index+1, len(infos)):
            intro = infos[x]
            if intro.startswith("◎"):
                break
            print(intro)
```
